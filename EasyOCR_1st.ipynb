```python
import easyocr
import cv2
import numpy as  np
from matplotlib import pyplot as plt
```

## IMAGE 1

```python
img_path = r"/content/surf.jpeg"
```

```python
reader = easyocr.Reader(['en'], gpu = "False")
result = reader.readtext(img_path)
print(result)
```

```python
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread(img_path)
img = cv2.rectangle(img,top_left,bottom_right,(0,255,0)  ,5)
img = cv2.putText(img, text, top_left, font , .5 , (255,255,255) , 2, cv2.LINE_AA)
plt.imshow(img)
plt.show()
```

## SAMPLE IMAGE 2
(Handling Multiple Lines)

```python
img_path1 = "/content/sign.png"
```

```python
reader1 = easyocr.Reader(['en'], gpu = "False")
result1 = reader1.readtext(img_path1)
print(result1)
```

```python
img1 = cv2.imread(img_path1)
spacer = 100
for detection in result1: 
    top_left = tuple(detection[0][0])
    bottom_right = tuple(detection[0][2])
    text1 = detection[1]
    img1 = cv2.rectangle(img1,top_left,bottom_right,(0,255,0),3)
    img1 = cv2.putText(img1,text1,(20,spacer), font, 0.5,(0,255,0),2,cv2.LINE_AA)
    spacer+=15
    
plt.imshow(img1)
plt.show()
```
